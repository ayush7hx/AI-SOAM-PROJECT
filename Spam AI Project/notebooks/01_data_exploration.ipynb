```python
# %% [markdown]
# # 1. Data Exploration - Spam SMS Dataset
# 
# **Objective:** Understand the dataset before building the model

# %% [code]
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/spam.csv', encoding='latin-1')
print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())

# %% [code]
# Check column names
print("Column names:", df.columns.tolist())

# Keep only relevant columns
df = df[['v1', 'v2']]
df.columns = ['label', 'message']
print("\nCleaned columns:", df.columns.tolist())

# %% [code]
# Check distribution
print("\nLabel Distribution:")
print(df['label'].value_counts())

# Visualize
plt.figure(figsize=(8, 5))
df['label'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Spam vs Ham Messages')
plt.xlabel('Message Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# %% [code]
# Check message lengths
df['message_length'] = df['message'].apply(len)
print("\nAverage message length:")
print(df.groupby('label')['message_length'].mean())

# %% [markdown]
# **Insights from Exploration:**
# 1. We have 5574 messages
# 2. 4825 are Ham (87%), 747 are Spam (13%)
# 3. Spam messages are typically longer than Ham messages
```
